# Generate 100 samples of different sizes N from a Chi-Squared, Normal, Poisson, Binomial, Logistic and Uniform distributions

```python
from __future__ import print_function
__author__="mmh"
import pylab as pl
import numpy as np
import os
%pylab inline

md = {}
myMean = 100
means = {}
mySize = 2000/(np.array(range(1,100)+[10]))
np.random.seed(101)
np.random.RandomState()
```

```python
#generating samples for chi-squared
md['chisq'] = {}
for n in mySize:
    md['chisq'][n] = np.random.chisquare(df=myMean, size=n)
means['chisq']={}

#make an empty figure
fig_mu_N = pl.figure(figsize=(10,10))
chisq_mu_N = fig_mu_N.add_subplot(111)

#calculate means for each sample and plot
for nn in md['chisq'].iterkeys():
    means['chisq'][nn] = md['chisq'][nn].mean()
    chisq_mu_N.plot(nn, means['chisq'][nn],'o')
    
#label axises and title
chisq_mu_N.set_xlabel('sample size')
chisq_mu_N.set_ylabel('sample mean')
chisq_mu_N.set_title('100 Chi-Squared Distributions')

#add solid black line that represents mean
chisq_mu_N.plot([min(mySize),max(mySize)],[myMean,myMean],'k')
```

Chi-squared distibutions converge to a normal distibution at higher degrees of freedom but skews right. It's hard to tell which way this skews, a histogram might be a better visualization. There should be more points above the mean line.

```python
#generating samples for normal
md['normal'] = {}
for n in mySize:
    md['normal'][n] = np.random.normal(myMean, 1, n)
means['normal'] = {}

#make an empty figure
fig_mu_N = pl.figure(figsize=(10,10))
normal_mu_N = fig_mu_N.add_subplot(111)
#calculate means for each sample and plot
for nn in md['normal'].iterkeys():
    means['normal'][nn] = md['normal'][nn].mean()
    normal_mu_N.plot(nn, means['normal'][nn],'o')
#label axises and title
normal_mu_N.set_xlabel('sample size')
normal_mu_N.set_ylabel('sample mean')
normal_mu_N.set_title('100 Normal Distributions')
#add solid black line that represents mean
normal_mu_N.plot([min(mySize),max(mySize)],[myMean,myMean],'k')
```

A normal distibution is symmetric around the mean. Again, hard to tell, but this looks symmetric.

```python
#generating samples for Poisson
md['poisson'] = {}
for n in mySize:
    md['poisson'][n] = np.random.poisson(myMean, n)
means['poisson'] = {}

#make an empty figure
fig_mu_N = pl.figure(figsize=(10,10))
poisson_mu_N = fig_mu_N.add_subplot(111)
#calculate means for each sample and plot
for nn in md['poisson'].iterkeys():
    means['poisson'][nn] = md['poisson'][nn].mean()
    poisson_mu_N.plot(nn, means['poisson'][nn],'o')
#label axises and title
poisson_mu_N.set_xlabel('sample size')
poisson_mu_N.set_ylabel('sample mean')
poisson_mu_N.set_title('100 Poisson Distributions')
#add solid black line that represents mean
poisson_mu_N.plot([min(mySize),max(mySize)],[myMean,myMean],'k')
```

Poisson distributions converge to binomial distibutions at large lambdas, with decreasing positive skewness. Hard to tell, but there should be more points above the mean line.

```python
#generating samples for binomial distribution
md['binomial'] = {}
for n in mySize:
    md['binomial'][n] = np.random.binomial(myMean/.5, .5, n)
means['binomial'] = {}

#make an empty figure
fig_mu_N = pl.figure(figsize=(10,10))
binomial_mu_N = fig_mu_N.add_subplot(111)
#calculate means for each sample and plot
for nn in md['binomial'].iterkeys():
    means['binomial'][nn] = md['binomial'][nn].mean()
    binomial_mu_N.plot(nn, means['binomial'][nn],'o')
#label axises and title
binomial_mu_N.set_xlabel('sample size')
binomial_mu_N.set_ylabel('sample mean')
binomial_mu_N.set_title('100 Binomial Distributions')
#add solid black line that represents mean
binomial_mu_N.plot([min(mySize),max(mySize)],[myMean,myMean],'k')
```

Binomial distributions with even odds are symmetrical around the mean. This looks more or less symmetrical.

```python
#generating samples for logistic distribution
md['logistic'] = {}
for n in mySize:
    md['logistic'][n] = np.random.logistic(myMean,100,n)
means['logistic'] = {}

#make an empty figure
fig_mu_N = pl.figure(figsize=(10,10))
logistic_mu_N = fig_mu_N.add_subplot(111)
#calculate means for each sample and plot
for nn in md['logistic'].iterkeys():
    means['logistic'][nn] = md['logistic'][nn].mean()
    logistic_mu_N.plot(nn, means['logistic'][nn],'o')
#label axises and title
logistic_mu_N.set_xlabel('sample size')
logistic_mu_N.set_ylabel('sample mean')
logistic_mu_N.set_title('100 Logistic Distributions')
#add solid black line that represents mean
logistic_mu_N.plot([min(mySize),max(mySize)],[myMean,myMean],'k')
```

Logistic distributions are symmetric but have fatter tails. We see some extreme values on both sides here.

```python
#generating samples for uniform distribution
md['uniform'] = {}
for n in mySize:
    md['uniform'][n] = np.random.uniform(myMean,myMean,n)
means['uniform'] = {}

#make an empty figure
fig_mu_N = pl.figure(figsize=(10,10))
uniform_mu_N = fig_mu_N.add_subplot(111)
#calculate means for each sample and plot
for nn in md['uniform'].iterkeys():
    means['uniform'][nn] = md['uniform'][nn].mean()
    uniform_mu_N.plot(nn, means['uniform'][nn],'o')
#label axises and title
uniform_mu_N.set_xlabel('sample size')
uniform_mu_N.set_ylabel('sample mean')
uniform_mu_N.set_title('100 Uniform Distributions')
#add solid black line that represents mean
uniform_mu_N.plot([min(mySize),max(mySize)],[myMean,myMean],'k')
```

This uniform distribution appears exactly along the mean line because the mean will not change with the size of sample if all the values are the same. I guess this wasn't a very interesting distribution.

```python
#plot histogram of all samples
allMeans = []
for n in means:
    for i in means[n]:
        allMeans.append(means[n][i])
pl.figure(figsize = (10,10))
pl.hist(allMeans,bins=50)
pl.xlabel('sample mean')
pl.ylabel('sample size')
```
